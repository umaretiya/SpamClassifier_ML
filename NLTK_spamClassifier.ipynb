```python
import os
import pandas as pd
import re
import nltk

from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split

from sklearn.naive_bayes import MultinomialNB
```

```python
path= "D:\\PycharmProjects\\iNeuron_ML_Projects\\SpamClassifier-master\\smsspamcollection"
file_name = "SMSSpamCollection"
file_path = os.path.join(path, file_name)
```

```python
dataset = pd.read_csv(file_path, sep='\t', header=None)
```

```python
dataset.columns =['label', 'message']
```

```python
nltk.download('stopwords')
```

```python
stopwords_list  = stopwords.words('english')
stopwords_list
```

```python
ps = PorterStemmer()
```

```python
corpus = []
for i in range(0, len(dataset)):
    review = re.sub('[^a-zA-Z0-9]',' ', dataset['message'][i])
    review = review.lower()
    review = review.split()
    review = [ps.stem(word) for word in review if not word in stopwords_list]
    review= ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
y = pd.get_dummies(dataset['label'], drop_first=True)
y
```

```python
X_train,X_test,y_train,y_test = train_test_split(corpus,y, test_size=0.2, random_state=42)
```

```python
dataset['label'].value_counts()
```

```python
cv = CountVectorizer(max_features=2500, ngram_range=(1, 2))
X_train_trans = cv.fit_transform(X_train).toarray()
```

```python
X_test_trans = cv.transform(X_test).toarray()
```

```python
X_train_trans.shape, X_test_trans.shape
```

```python
cv.vocabulary_
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier()
classifier.fit(X_train_trans, y_train)
```

```python
y_pred = classifier.predict(X_test_trans)
```

```python
from sklearn.metrics import classification_report, accuracy_score,confusion_matrix
```

```python
accuracy_score(y_test, y_pred)
```

```python
confusion_matrix(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```
